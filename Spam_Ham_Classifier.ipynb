```python
import nltk
from nltk.corpus import stopwords
from nltk.stem import PorterStemmer
from nltk.stem import WordNetLemmatizer
from nltk import word_tokenize
nltk.download('stopwords')
nltk.download('wordnet')
import re

lemmatizer=WordNetLemmatizer()
stemmer=PorterStemmer()
stop_words = set(stopwords.words('english'))
```

```python
def preprocess_text(text):
    text=text.lower()
    text=re.sub(r"[^A-Za-z]"," ",text).split() #Removing every thing other than alphatbet thing like !,?,etc
    text=[word for word in text if word not in stop_words] #Removing stopwords
    text=[lemmatizer.lemmatize(word) for word in text] #Using lemmatizer
    text=[stemmer.stem(word) for word in text] #Using stemmer to get stem of the word
    text=" ".join(text)
    return text
```

```python
df["cleaned_text"]=df.text.apply(preprocess_text)
```

```python
df.head()
```
